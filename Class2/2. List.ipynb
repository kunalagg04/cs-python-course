# List

Lists are just like the arrays, declared in other languages.
#### A single list may contain DataTypes like Integers, Strings, as well as Objects.
####  Lists are also very useful for implementing stacks and queues. 
#### Lists are mutable, and hence, they can be altered even after their creation.

```python
List = [] 
print("Intial blank List: ") 
print(List) 
  
```

```python
# Creating a List with  
# the use of a String 
List = ['Shastra'] 
print("\nList with the use of String: ") 
print(List) 
  
```

```python
# Creating a List with 
# the use of multiple values 
List = ["Coding", "is", "love"] 

print(List)
 
```

```python
# Creating a Multi-Dimensional List 
# (By Nesting a list inside a List) 
List = [['Are', 'you'] , ['crazy']] 
print(List) 
```

```python
#Creating a List with  
# the use of Numbers 
# (Having duplicate values) 
List = [1, 2, 4, 4, 3, 3, 3, 6, 5] 

print(List) 
  
```

```python
# Creating a List with  
# mixed type of values 
# (Having numbers and strings) 
List = [1, 2, 'Geeks', 4, 'For', 6, 'Geeks'] 
 
print(List) 
```

### Adding elements to list

#### Append() method
It adds elements at the end of list

```python

List = [] 
  
# Addition of Elements  
# in the List 
List.append(1) 
List.append(2) 
List.append(4) 

print(List) 
```

```python
# Adding Tuples to the List 
List.append((5, 6)) 

print(List) 
```

```python
# Addition of List to a List 
List2 = ['For', 'Geeks'] 
List.append(List2) 

print(List)
```

#### Insert() method

Adds elements at desired position . Takes two args.

```python
# Addition of Element at specific Position 
# (using Insert Method) 
List=[0,1,2]
List.insert(3, 12) 
print(List) 
```

### Accessing elements from the List
 

```python
# accessing a element from the  
# list using index number 
List=[0,1,2,3,4]
print("Accessing a element from the list") 
print(List[0])  
print(List[2]) 
```

```python
# Creating a Multi-Dimensional List 
# (By Nesting a list inside a List) 
List = [['coder', 'or'] , ['decoder']] 
  
# accessing a element from the  
# Multi-Dimensional List using 
# index number 
print("Acessing a element from a Multi-Dimensional list") 
print(List[0][1]) 
print(List[1][0]) 
```

```python
#Negative indexing
List=[0,1,2]
# print the last element of list 
print(List[-1]) 
  
# print the third last element of list  
print(List[-3]) 
```

# Removing element
  
####  Pop() function can also be used to remove and return an element from the set, but by default it removes only the last element of the set, to remove element from a specific position of the List, index of the element is passed as an argument to the pop() method. Returns removed element. 

```python
# Removing element from the list using the pop() method 
List=[0,1,2,3]
print(List.pop()) 
print(List) 
```

```python
# Removing element at a specific location from the list using the pop() method 
List=[0,1,2,3]
List.pop(2) 
print("\nList after popping a specific element: ") 
print(List) 
```

## Slicing of a List
Slicing doesn't modify the current list .


```python
List=['p','y','t','h','o','n']
# Print elements of a range using Slice operation 
Sliced_List = List[3:8] 
print("Slicing elements in a range 1-5: ") 
print(Sliced_List) 
```

```python
# Print elements from a pre-defined point to end 
Sliced_List = List[2:] 
print(Sliced_List) 
```

```python
# Print elements from beginning to a pre-defined point using Slice 
Sliced_List1 = List[:-3] 
print("Elements sliced till 6th element from last: ") 
print(Sliced_List1) 
```

```python
# Printing elements in reverse using Slice operation 
Sliced_List = List[::-1] 
print(Sliced_List)
```
