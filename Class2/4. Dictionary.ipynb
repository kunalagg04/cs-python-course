# Dictionary

Dictionary in Python is an unordered collection of data values, used to store data values like a map, which unlike other Data Types that hold only single value as an element.
### Dictionary holds key:value pair. Key value is provided in the dictionary to make it more optimized. 
Each key-value pair in a Dictionary is separated by a colon :, whereas each key is separated by a ‘comma’.

### Values in a dictionary can be of any datatype and can be duplicated, whereas keys can’t be repeated and must be immutable.


```python
# Creating a Dictionary with Integer Keys 
Dict = {1: 'Hustle', 2: 'For', 3: 'success'} 
print(Dict)
```

```python
# Creating a Dictionary with dict() method 
Dict = dict({1: 'Hustle', 2: 'For', 3: 'success'} ) 
print(Dict) 
```

```python
# Creating a Dictionary with each item as a Pair 
Dict = dict([(1, 'good'), (2, 'going')]) 
print(Dict) 
```

## Nested Dictionary


```python
# Creating a Nested Dictionary
Dict = {1: 'Coding', 2: 'Shastra',  
        3:{'A' : 'Welcome', 'B' : 'Boi'}} 
  
print(Dict)  
```

## Adding elements 
 While adding a value, if the key value already exists, the value gets updated otherwise a new Key with the value is added to the Dictionary.

```python
# Creating an empty Dictionary 
Dict = {}  
  
# Adding elements one at a time 
Dict[0] = 'rofl'
Dict[2] = 'lol'
Dict[3] = 1

print(Dict) 
```

```python
# Adding set of values to a single Key 
Dict['Value_set'] = 2, 3, 4
print(Dict) 
```

## Accessing elements from a Dictionary


```python
# Creating a Dictionary  
Dict = {1: 'time', 'name': 'is', 3: 'money'} 

# accessing a element using key 
print(Dict['name']) 
```

```python
# accessing a element using get() 
print(Dict.get(3)) 
```

## Removing Elements from Dictionary


```python
# Initial Dictionary 
Dict = { 5 : 'Welcome', 6 : 'To', 7 : 'Delhi', 
        'A' : {1 : 'coding', 2 : 'is', 3 : 'love'}, 
        'B' : {1 : 'Coder', 2 : 'Life'}} 
print(Dict) 
```

```python
# Deleting a Key value 
del Dict[6] 
print(Dict) 
```

```python
# Deleting a Key from Nested Dictionary 
del Dict['A'][2] 
print(Dict)
```

```python
# Deleting a Key using pop() 
Dict.pop(5) 
print(Dict) 
```
