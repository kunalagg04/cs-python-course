# Tuples
The important difference between a list and a tuple is that tuples are immutable.
Tuples can contain any number of elements and of any datatype (like strings, integers, list, etc.). 

```python
# Creating a tuple 
Tuple1 = (1,2,3,4) 
print (Tuple1) 
  
```

```python
#Creating a tuple with help of inbuilt function
Tuple1 = tuple("bla") 
print(Tuple1) 
```

## Concatenation of Tuples

Note- Only same datatypes can be combined with concatenation, an error arises if a list and a tuple are combined.


```python
# Concatenaton of tuples 
Tuple1 = (0, 1, 2, 3) 
Tuple2 = ('Geeks', 'For', 'Geeks') 
  
Tuple3 = Tuple1 + Tuple2 
print(Tuple3)
```

## Slicing of Tuple

Slicing of a Tuple is done to fetch a specific range or slice of sub-elements from a Tuple.

```python
tuple=('h','u','s','t','l','e')
tuple1=tuple[1:5] 
#Note that elem. at index 5 will not be included
print(tuple1)
```

## Deleting a Tuple

Tuples are immutable and hence they do not allow deletion of a part of it. Entire tuple gets deleted by the use of del() method.

```python
# Deleting a Tuple 
  
Tuple1 = (0, 1, 2, 3, 4) 
del Tuple1 
  
print(Tuple1) 
```
