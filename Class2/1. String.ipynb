# Strings

Strings are arrays of characters.
Strings in Python can be created using single quotes or double quotes or even triple quotes.

```python
# Creating a String with single Quotes 
String1 = 'Hi Bro'
print("String with the use of Single Quotes: ") 
print(String1) 
```

## Accessing characters in Python

#### Individual characters of a String can be accessed by using the method of Indexing.
While accessing an index out of the range will cause an IndexError. Only Integers are allowed to be passed as an index, float or other types will cause a TypeError.



#### To access a range of characters in the String, method of slicing is used.

```python
String1="Shastra"
# Printing First character 
print("\nFirst character of String is: ") 
print(String1[0]) 
```

```python
# Printing Last character 
print("\nLast character of String is: ") 
print(String1[-1]) 
```

#### In slicing the last index of range is not included . For example if you input string[1:3] , then char at index 3 is not included.

```python
# Printing 3rd to 12th character 
print("Slicing characters: ") 
print(String1[3:6]) 
```

#### Updation and deletion of characters of string is not allowed.
#### We can update and delete(using del keyword) whole string though.

```python
a="blabla"
#updating string char
a[0]=a
#it will give error
```

```python
#deleting string char
del(a[2])
#it will give error
```

```python
#updating string
a="blah"
```

```python
#deleting string
del(a)
print(a)
```

### Escape Sequencing in Python
Using single quotes in a string wrapped inside single quotes creates error . 
#### So, we use / to avoid error .

```python
#Escaping Single Quote  
String1 = 'I\'m a "Geek"'
print("Escaping Single Quote: ") 
print(String1)
```
